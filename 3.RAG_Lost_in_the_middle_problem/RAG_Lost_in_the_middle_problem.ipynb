### **Load Embedding model to convert Data to Vectors**

```python
from langchain.embeddings import HuggingFaceEmbeddings,HuggingFaceBgeEmbeddings
```

```python
hf_embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
hf_bge_embeddings=HuggingFaceBgeEmbeddings(model_name="BAAI/bge-large-en")
```

### **Now ingest the data to chromadb using embeddings**

```python
from langchain.vectorstores import Chroma
import chromadb
```

```python
import os
os.getcwd()
```

```python
CURRENT_DIR = os.path.dirname(os.path.abspath("."))
```

```python
DB_DIR = os.path.join(CURRENT_DIR,"/content/db")
```

```python
DB_DIR
```

```python
client_settings = chromadb.config.Settings(
    is_persistent=True,
    persist_directory=DB_DIR,
    anonymized_telemetry=False
)
```

```python
attention_vectorstore=Chroma.from_documents(
    text_attention,
    hf_bge_embeddings,
    client_settings=client_settings,
    collection_name="attention",
    collection_metadata={"hnsw":"cosine"},
    persist_directory="/store/attention"
)
```

```python
ULM_vectorstore=Chroma.from_documents(
    text_attention,
    hf_bge_embeddings,
    client_settings=client_settings,
    collection_name="ULM",
    collection_metadata={"hnsw":"cosine"},
    persist_directory="/store/ULM"
)
```

### **Create Retriever on top of the vectorstore**

```python
retriever_attention=attention_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k":5,"include_metadata":True})
```

```python
retriever_ULM=ULM_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k":5,"include_metadata":True})
```

### **Lets Merge both retriever**

```python
from langchain.retrievers.merger_retriever import MergerRetriever
```

```python
retriever_merged = MergerRetriever(retrievers=[retriever_attention,retriever_ULM])
```

```python
for chunks in retriever_merged.invoke("What is finetuning?"):
  print(chunks.page_content)
```

Since the result is too much messy now lets fine it according to question and overcome the solution of lost io the middle

### **Create a Pipeline for LLM**

```python
from langchain.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter,
    LongContextReorder
)
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.retrievers import ContextualCompressionRetriever
```

```python
from re import search
filter = EmbeddingsRedundantFilter(embeddings=hf_bge_embeddings)
reordering = LongContextReorder()
pipeline = DocumentCompressorPipeline(transformers=[filter,reordering])
compression_retriever_reorderd = ContextualCompressionRetriever(
    base_compressor=pipeline,base_retriever=retriever_merged,
    search_kwargs={"k":3,"include_metadata":True}
)
```

```python
docs=compression_retriever_reorderd.invoke("What is Attention?")
print(len(docs))
```

```python
print(docs[0].page_content)
```

```python
from langchain.llms import LlamaCpp
```

```python
llms=LlamaCpp(
    straming=True,
    model_path="/content/drive/MyDrive/zephyr-7b-beta.Q4_0.gguf",
    max_tokens=1500,
    temperature=0.75,
    top_p=1,
    gpu_layers=0,
    stream=True,
    verbose=True,
    n_threades=int(os.cpu_count()/2),
    n_ctx=4096
)
```

```python
from langchain.chains import RetrievalQA
```

```python
qa=RetrievalQA.from_chain_type(
    llm=llms,
    chain_type="stuff",
    retriever=compression_retriever_reorderd,
    return_source_documents=True
)
```

```python
query="what is attention?"
result=qa(query)
print(result["result"])
```
